# Missing Data

NAN: Not a Number

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,4,np.nan,9],
             'B':[3,np.nan,1,np.nan],
             'C':[6,7,8,9],
             'D':['Pass',np.nan,np.nan,'Fail']})
df
```

```python
df.dropna()  #default value of axis is 0
```

```python
df.dropna(axis=1)
```

```python
df
```

```python
df.dropna(thresh = 3, axis = 1)
```

```python
df.fillna(value='Missing Value')
```

```python
df['A'].fillna(0)
```

```python
df.fillna({'A':2,'B':'Nothing'})
```

```python
df['A'].fillna(value=df['C'].mean())
```

```python
df
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='ffill', axis = 1)
```

```python
df.fillna(method='bfill')
```

```python
df
```

```python
df.interpolate()
```

```python
df1 = pd.DataFrame({'A':[1,np.nan,np.nan,9],
             'B':[3,np.nan,1,np.nan],
             'C':[6,7,8,9],
             'D':['Pass',np.nan,np.nan,'Fail']})
df1
```

```python
df1.interpolate()
```

# Special Values

```python
df3 = pd.DataFrame({'A':[1,2,np.nan,7777],
             'B':[3,np.nan,1,np.nan],
             'C':[6,7,9999,9],
             'D':['Pass',np.nan,np.nan,'Fail']})
df3
```

```python
df3.replace(9999,np.nan)
```

```python
df3.replace([9999,7777],np.nan)
```

```python
df4 = df3.replace(7777,"Pass")
df4
```

```python
df4.replace({'A':'Pass','C':9999},np.nan)
```

```python
df4.loc[df4['C'] > 6, 'C'] = 0
```

```python
df4
```

```python
df4.replace({'D':'Fail',
            'B':np.nan,
            'C':0},5)
```

```python
#Replace Entire List

st = pd.DataFrame({'name':['riya','mia','neha','swati'],
             'score':['E','P','P','A']})
st
```

```python
st.replace(['E','P','A'],['Excellent','Poor','Average'])
```
