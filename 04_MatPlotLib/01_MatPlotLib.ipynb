# Basic MatPlotLib Commands

```python
plt.plot(x,y, 'green')
plt.xlabel('Put X axis title here')
plt.ylabel('Put Y axis title here')
plt.title('Put Title for this figure')
plt.legend(labels='x')
plt.show()
```

# Object Oriented Method

```python
fig = plt.figure() #empty canvas

axes = fig.add_axes([1,0.1,0.8,0.8]) 

axes.plot(x,y, 'r')
axes.set_xlabel('Set X label here')
axes.set_ylabel('Set y label here')
axes.set_title('Set Title')
plt.show()
#left, bottom, width, height
```

```python
#Multi-Plots
fig = plt.figure() #empty canvas

axes1 = fig.add_axes([0,0,0.5,0.8])
axes2 = fig.add_axes([0.6,0,0.5,0.8])
#left, bottom, width, height
axes1.plot(x,y, 'r')
axes1.set_xlabel('Set X1 label here')
axes1.set_ylabel('Set y1 label here')
axes1.set_title('Set Axes 1 Title')

axes2.plot(y,x, 'b')
axes2.set_xlabel('Set X2 label here')
axes2.set_ylabel('Set y2 label here')
axes2.set_title('Set Axes 2 Title')

plt.show()
```
