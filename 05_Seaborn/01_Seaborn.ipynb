# distplot

shows distribution of a univariate(one variable) set of observations. 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df['total_bill'])
```

kde: kernel density estimation calculates the density of features in the neighbourhood around those features. 

```python
sns.distplot(df['total_bill'],kde=True, bins = 30)

#bins define smoothness or sharpness in your plot
```

# jointplot

match up two distplots for bivariate data. kind parameter: 'scatter', 'reg','kde','hex','resid'

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='kde')
plt.show()
```

# Pairplot

pairwise relationships across entire df (for numerical column, although it supports categorical columns through 'hue')

```python
df.columns
```

```python
sns.pairplot(df[['total_bill', 'tip', 'size',
       'price_per_person']])
```

```python
sns.pairplot(df, hue = 'day', palette="viridis")
```

# rugplot

dash mark for every point on a univariate distribution. Building block of a KDE plot

```python
sns.kdeplot(df['tip'])
sns.rugplot(df['tip'])
```

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```

# Categorical Data Plots


### Bar Plot and Count Plot

```python
sns.barplot(x='day',y='tip',data=df, hue = 'sex', estimator=np.mean)
```

```python
sns.countplot(data=df, x='smoker', saturation = 0.5, hue = 'day')
```

### Boxplot and Violinplot

```python
sns.boxplot(y='smoker',x='total_bill',data=df,medianprops={"color": "white"},orient="h")
```

```python
sns.violinplot(x='smoker',y='total_bill', data=df, hue='sex', split = True)
```

# Stripplot

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
#swarmplot
sns.swarmplot(x='day',y='total_bill',data=df)
```
