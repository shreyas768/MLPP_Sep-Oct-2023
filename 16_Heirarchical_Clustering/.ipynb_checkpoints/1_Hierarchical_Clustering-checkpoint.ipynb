```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop_duplicates()
```

```python
df_w_dum = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_w_dum
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
#help(MinMaxScaler)
```

```python
scaled_data = scaler.fit_transform(df_w_dum)
```

```python
scaled_data
```

```python
sc_df = pd.DataFrame(scaled_data,columns=df_w_dum.columns)
```

```python
sc_df.describe()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(sc_df)
```

```python
sns.clustermap(sc_df)
```

```python
sns.clustermap(sc_df, row_cluster=False)
```

```python
sns.clustermap(sc_df, col_cluster=False)
```

# Using SkLearn

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
mod = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = mod.fit_predict(sc_df)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df,x='acceleration',y='mpg', hue=cluster_labels)
```
