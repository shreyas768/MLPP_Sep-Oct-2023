# DEF for label determination

```python
def disp_cat(model,ds):
    pred = model.fit_predict(ds)
    sns.scatterplot(data=ds, x = 'X1', y = 'X2', hue = pred, palette='Set1')
```

# KMeans

```python
from sklearn.cluster import KMeans
model  = KMeans(n_clusters=2)
```

```python
disp_cat(model,moons)
```

```python
model1  = KMeans(n_clusters=3)
disp_cat(model1,blobs)
```

```python
disp_cat(model,circles)
```

# DBSCAN Results

```python
from sklearn.cluster import DBSCAN

mod_db = DBSCAN(eps= 0.6)
```

```python
#help(DBSCAN)
```

```python
disp_cat(mod_db,blobs)
```

```python
mod_db1 = DBSCAN(eps=0.2)
disp_cat(mod_db1,moons)
```
